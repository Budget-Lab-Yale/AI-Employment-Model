```python
import os
import glob
from datetime import datetime
from pathlib import Path
from ipumspy import IpumsApiClient, MicrodataExtract, readers, ddi
from pandas import read_csv

from parsing     import build_linked_dataset
```

```python

# Should replace with your own key
IPUMS_API_KEY = "59cba10d8a5da536fc06b59d3b7781dc64fe4a1ea7c19179c8a6e41b"
ipums = IpumsApiClient(IPUMS_API_KEY)

stamp = datetime.now().strftime('%Y%m%d%H')

extract = MicrodataExtract(
    collection = "cps",
    description = "Test CPS Extract",
    # Visit this site and select the appropriate code(s):
    # https://cps.ipums.org/cps-action/samples/sample_ids
    samples = ["cps2025_11s"],
    #samples = samples,
    variables = [
        "YEAR",     # Survey year
        "SERIAL",   # Household serial number
        "MONTH",    # Month
        "HWTFINL",  # Household weight, basic monthly
        "CPSID",    # CPSID, household record
        "PERNUM",   # Person number in sample unit
        "WTFINL",   # Final Basic Weight
        "CPSIDP",   # CPSID, person record
        "CPSIDV",   # Validated Longitudinal Identifier
        "AGE",      # Age
        "SEX",      # Sex
        "EMPSTAT",  # Employment status
        "LABFORCE", # Labor force status
        "OCC",      # Occupation
        "OCC2010",  # Occupation, 2010 basis
        "IND",      # Industry
        "DURUNEMP", # Continuous weeks unemployed
        "EDUC",     # Educational attainment record
    ]
)

# Submit the extract request
ipums.submit_extract(extract)
print(f"Extract submitted with id {extract.extract_id}")

# Wait for the extract to finish
ipums.wait_for_extract(extract)

# Download the extract
DOWNLOAD_DIR = Path("/gpfs/gibbs/project/sarin/shared/raw_data/CPS-Monthly/" + str(stamp))
DOWNLOAD_DIR.mkdir(parents=True, exist_ok=True)
ipums.download_extract(extract, download_dir=DOWNLOAD_DIR)

os.rename(
    "/gpfs/gibbs/project/sarin/shared/raw_data/CPS-Monthly/" + str(stamp) + "/cps_000" + str(extract.extract_id) + ".dat.gz", 
    "/gpfs/gibbs/project/sarin/shared/raw_data/CPS-Monthly/" + str(stamp) + "/cps.dat.gz"
)
os.rename(
    "/gpfs/gibbs/project/sarin/shared/raw_data/CPS-Monthly/" + str(stamp) + "/cps_000" + str(extract.extract_id) + ".xml", 
    "/gpfs/gibbs/project/sarin/shared/raw_data/CPS-Monthly/" + str(stamp) + "/cps.xml"
)
```

```python
# Get the DDI
ddi_file = list(DOWNLOAD_DIR.glob("*.xml"))[0]
ddi = readers.read_ipums_ddi(ddi_file)

# Get the data
cps_data = readers.read_microdata(ddi, DOWNLOAD_DIR / ddi.file_description.filename)

# Now you can do your analysis
print(cps_data.head())
```
